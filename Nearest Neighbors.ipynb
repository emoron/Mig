Let P be a set of  points in the plane, Q be a set o m  points in the plane, and k  be an integer.

For each element $q \in Q$  , you want to find the k points in P with the minimum distance to q .

Input Format

The first line contains three space separated integers,n ,m  and k.

The next n lines contain each a point of P, given as two space separated integers with coordinates  and .

The next m lines contain points in Q, given in the same fashion.

Constraints

1 <= k <= 10
k <= n <= 10^5
k <= m <= 10^5




It is ensured that for each point in Q, there is a unique set of k nearest points.

For 40% of the score, n*m <= 2* 10^5.

Output Format

For each point q, output k lines, each with the coordinates of the k nearest points, in a new line. The points must be ordered in terms of distance to q (ie print the points closest to q first)

Sample Input 0

2 1 2
0 0
1 0
0 1
Sample Output 0


0 0
1 0
Sample Input 1

3 1 1
0 0
1 0
0 1
1 2
Sample Output 1

0 1
Explanation 1

(0,1) is the closest point to (1,2).

Sample Input 2

2 1 2
0 0
1 1
2 2
Sample Output 2

1 1
0 0
Explanation 2

We have to print (1,1) first, since it's closer to the query point (2,2).
